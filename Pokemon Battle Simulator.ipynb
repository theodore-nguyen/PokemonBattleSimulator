#### The idea of this project is to create a simluator in which you 'a user' can choose between 3 starting pokemon and a battle random 'enemy' pokemon to gain xp. 

#### Certain elements counters other certain elements like 'Fire' counters 'Grass', 'Water' counters 'Fire', and 'Grass' counters 'Water'

#### However, the enemy pokemon will attack when you attack. The end goal is to evolve your pokemon to the max evolution without dying.

#### As well as to tell the user to try again, if an incorrect input is made.

### The code for the pokemon battle simulator is below:


```python
import color


#Below are stored strings for easier formatting
c_text = color.BOLD + color.RED + "  Charmander  " + color.END
c2_text = color.BOLD + color.RED + "  Charmeleon  " + color.END
c3_text = color.BOLD + color.RED + "  Charizard  " + color.END

b_text = color.BOLD + color.GREEN + "  Bulbasaur  " + color.END
b2_text = color.BOLD + color.GREEN + "  Ivysaur  " + color.END
b3_text = color.BOLD + color.GREEN + "  Venusaur  " + color.END

s_text = color.BOLD + color.BLUE + "  Squirtle   " + color.END
s2_text = color.BOLD + color.BLUE + "  Wartortle  " + color.END
s3_text = color.BOLD + color.BLUE + "  Blastoise  " + color.END

congratz = color.BOLD + color.GREEN + "Congratulations!!" + color.END

class Pokemon():
    
    """
    Parent Class for all pokemon classes
    
    """
    def __init__(self, hp = None, element = None):
        
        self.hp = hp
        self.element = element
        
class Evil(Pokemon):
    """
     Evil class that inherits the Pokemon Class
     This is the parent class to the specific 'element' enemys
     
    """
    
    def __init__(self, hp, xp, element = None, counter = None):
        
        super().__init__(hp, element)
        self.counter = counter
        self.xp = xp
        
    def tackle(self, user):
        """
        This is the tackle method used by the enemy class onto the user's starter pokemon
        User's Pokemon loses 25 health points
        """
        
        print(color.BOLD + "The enemy used tackle!" + color.END)
        user.hp -= 25

        
class GrassEnemy(Evil):
    """
    Child Class to Evil
    This is the enemy class that is created with the element 'Grass'
    """
    
    def __init__(self, hp, xp, element = 'Grass', counter = 'Fire'):
        
        super().__init__(hp, xp, element, counter)       
        
        
class FireEnemy(Evil):
    """
    Child Class to Evil
    This is the enemy class that is created with the element 'Fire'
    """
    
    def __init__(self, hp, xp, element = 'Fire', counter = 'Water'):
        
        super().__init__(hp, xp, element, counter)   
        
        
class WaterEnemy(Evil):
    """
    Child Class to Evil
    This is the enemy class that is created with the element 'Water'
    """
    
    def __init__(self, hp, xp, element = 'Water', counter = 'Grass'):
        
        super().__init__(hp, xp, element, counter)
       
        
#Starter Pokemon
class Starter(Pokemon):
    """
    This is the Parent class to the specific element of the User's choice of pokemon
    """
    def __init__(self, hp = 0, element = None, xp = 0):
        
        super().__init__(hp, element)
        self.xp = xp
        
    def tackle(self, enemy):
        """
        This is the tackle method used by the User onto the Enemy class
        Enemy loses 50 health points
        """
        
        print(color.BOLD + "Okayyy, hitem with the basic attack \n    Enemy took 50 dmg" + color.END)
        enemy.hp -= 50
        
    def special_attack(self, enemy):
        """
        This is the special ability used by the User onto the Enemy class 
        -----------------------------------------------------------------------
        
        This ability does depends on the User's element and the Enemy's Element
        
        Fire > Grass | Grass > Water | Water > Fire
        
        If it's a critical hit, enemy loses 150 health points
        If it's same element, enemy loses 100 health poitns
        
        """
                        
        if (self.element == enemy.counter):
            
            print(color.BOLD + "Critical Hit, NICEEEEE \n    Enemy took 150 dmg" + color.END)
            enemy.hp -= 150
            
        elif (self.element == enemy.element):
            
            print(color.BOLD + "Ehh, ok dmg i guess \n     Enemy took 100 dmg" + color.END)
            enemy.hp -= 100
            
        else: 
            
            print(color.BOLD + "Not very effective aint it, rather u use tackle u twad \n     " +
                  "Enemy took 25 dmg" + color.END)
            enemy.hp -= 25
            
    def heal(self, counter):
        """
        This is the heal method
        ---------------------------
        
        Only usable (3) times 
        
        Brings User's Starter Pokemon's Health back to Full HP(300)
        
        """
        
        if (counter <= 3):
            
            self.hp = 300
            print(color.BOLD + color.GREEN + "Your Pokemon has been healed!" + color.END +
                 "       Uses Left: " + color.BOLD + str(3 - counter) + color.END)
        else:
            
            print(color.BOLD + color.RED + "You have ran out of uses!" + color.END)
        
class Bulbasaur(Starter):
    """
    This is the child class of Starter
    This starter's element is Grass
    """

    def __init__(self, name = b_text, hp = 300, element = 'Grass', xp = 0):
        
        super().__init__(hp, element, xp)
        self.name = name
        
    def update_name(self):
        """
        Updates the name of the starter when it evolves
        
        The starter evolves first when xp >= 50 and 
        second when xp >= 150 
        
        This makes it easier to test for the winning condition (final evolution)
        """
        
        if (self.xp < 50):
            
            self.name = b_text
            
        elif (self.xp >= 50 and self.xp < 150):
            
            if (self.name != b2_text):
                
                print("Your " + b_text + "has evolved into " + b2_text)
                print(congratz)
                self.name = b2_text
            
        elif (self.xp >= 150):
            
            if (self.name != b3_text):      
                
                print("Your " + b2_text + "has evolved into " + b3_text)
                self.name = b3_text
                
class Charmander(Starter):
    """
    This is the child class of Starter
    This starter's element is Fire
    """

    def __init__(self, name = c_text, hp = 300, element = 'Fire', xp = 0):
        
        super().__init__(hp, element, xp)
        self.name = name
    
    def update_name(self):
        """
        Updates the name of the starter when it evolves
        
        The starter evolves first when xp >= 50 and 
        second when xp >= 150 
        
        This makes it easier to test for the winning condition (final evolution)
        """
        
        if (self.xp < 50):
            
            self.name = c_text
            
        elif (self.xp >= 50 and self.xp < 150):
            
            if(self.name != c2_text):
                
                print("Your " + c_text + "has evolved into " + c2_text)
                print(congratz)
                self.name = c2_text
                
        elif (self.xp >= 150):
            
            if(self.name != c3_text):
                
                print("Your " + c2_text + "has evolved into " + c3_text)
                self.name = c3_text
    
class Squirtle(Starter):
    """
    This is the child class of Starter
    This starter's element is Water
    """

    def __init__(self, name = s_text, hp = 300, element = 'Water', xp = 0):
        
        super().__init__(hp, element, xp)
        self.name = name 
    
    def update_name(self):
        """
        Updates the name of the starter when it evolves
        
        The starter evolves first when xp >= 50 and 
        second when xp >= 150 
        
        This makes it easier to test for the winning condition (final evolution)
        """
        
        if (self.xp < 50):
            
            self.name = s_text
            
        elif (self.xp >= 50 and self.xp < 150):
            
            if (self.name != s2_text):
                
                print("Your " + s_text + "has evolved into " + s2_text)
                print(congratz)
                self.name = s2_text
            
        elif (self.xp >= 150):
            
            if (self.name != s3_text):
                
                print("Your " + s2_text + "has evolved into " + s3_text)
                self.name = s3_text
            
```

```python
import random 

def run(): 
    """
    This is the main engine of the Pokemon Battle Simulator
    
    Takes in no parameters
    """
    
    #prompts user for their choice of starter pokemon
    
    pokemon_choice = input(color.BOLD + "Please choose your starter: " +
                          c_text +
                          b_text + 
                          s_text +  "\n")
    
    #checks the pokemon choice if its one of the 3 starter pokemons
    user = check_starter(pokemon_choice)
    
    alive = True
    
    #this counter is for how many heals the user can use, the max is '3'
    counter = 1
    
    while alive:
        
        #creates a random enemy
        enemy = random_enemy()
        
        #this is the battle phase between the user and the enemy pokemon
        #the user has two moves "tackle" and "special" and 3 heals
        battle_phase(user, enemy, counter)
        
        #updates after every battle to check whether the pokemon evolved
        user.update_name()
        
        #checks for the losing condition and the winning condition
        if (user.xp >= 150):
            print(color.BOLD + color.GREEN + "CONGRATULATIONSSSS!!!!! \n"+
                 "You have beat the game" + color.END)
            alive = False
        if (user.hp <= 0):
            print(color.BOLD + color.RED + "YOU HAVE DIED \nGAME OVER!" + color.END)
            alive = False
            
        
def check_starter(starter_choice):
    """
    Input Parameter
    -----------------
    User Input as a String 
    
    
    Output Parameter
    -------------------
    Returns the starter Pokemon object 
    dependent on User Input 

    """
  
    if starter_choice.lower() == 'charmander':
        starter = Charmander()
        print(color.BOLD + "You have chosen" + c_text)
    
    elif starter_choice.lower() == 'squirtle':
        starter = Squirtle()
        print(color.BOLD + "You have chosen" + s_text)
    
    elif starter_choice.lower() == 'bulbasaur':
        starter = Bulbasaur()
        print(color.BOLD + "You have chosen " + b_text)
    else:
        starter = None
        print(color.RED + 'Incorrect Response, Please choose again \n' + color.END)
        pokemon_choice = input(color.BOLD + "Please choose your starter: " +
                          c_text +
                          b_text + 
                          s_text +  "\n")
        check_starter(pokemon_choice)
    return starter
#creates a random enemy for user to fight and gain experience         
def random_enemy():
    """
    Creates a random enemy every battle for the user to fight
    
    Element is randomized 
    
    Health is randomized
    
    Xp is correlated with health (xp = 1/10 (hp))
    
    After choosing an element, this method randomly chooses from the list of pokemon names with that respective element
    for the User to fight
    
    """
    rand_int = random.randint(1,3)
    rand_xp = random.randint(1,5) * 10
    rand_hp = rand_xp * 10
    
    if (rand_int == 1):
        enemy_pokemon = FireEnemy(rand_hp, rand_xp)
        #chooses a random pokemon from the list with attribute fire
        enemy_name = random.choice(['Vulpix', 'Ninetales', 'Growlithe', 'Arcanine', 'Ponyta', 'Rapidash', 'Magmar',
                                   'Flareon', 'Moltres'])
        print(color.BOLD + "You have encountered a random enemy:   " + 
              color.RED + enemy_name + color.END)
        
    elif (rand_int == 2):
        enemy_pokemon = GrassEnemy(rand_hp, rand_xp)
        #chooses a random pokemon from the list with attribute grass
        enemy_name = random.choice(['Oddish', 'Gloom', 'Vileplume', 'Paras', 'Parasect', 'Bellsprout', 'Weepinbell',
                                   'Victreebel', 'Exeggcute', 'Exeggutor', 'Tangela'])
        
        print(color.BOLD + "You have encountered a random enemy:   " + 
              color.GREEN + enemy_name + color.END)
        
    else: 
        enemy_pokemon = WaterEnemy(rand_hp, rand_xp)
        #chooses a random pokemon from the list with attribute water
        enemy_name = random.choice(['Psyduck', 'Golduck', 'Poliwag', 'Poliwhirl', 'Poliwrath', 'Tentacool', 'Tentacruel',
                                   'Slowpoke', 'Slowbro', 'Seel', 'Dewgong', 'Shellder', 'Cloyster', 'Krabby', 'Kingler',
                                   'Horsea', 'Seadra', 'Goldeen', 'Seaking', 'Staryu', 'Starmie', 'Magikarp', 'Gyarados',
                                   'Lapras', 'Vaporeon', 'Omanyte', 'Omastar', 'Kabuto', 'Kabutops'])
        
        print(color.BOLD + "You have encountered a random enemy:   " + 
              color.BLUE + enemy_name + color.END)
        
    print(color.RED + "ENEMY HP: " + str(rand_hp))
    return enemy_pokemon

def battle_phase(user, enemy, counter):
    """
    Input Parameters
    ------------------------------
    
    User's Starter as Pokemon Object
    
    Enemy pokemon as Pokemon Object
    
    Counter as an Int 
        in order to keep track of how many heals are left
        
    """
    
    while (enemy.hp > 0 and user.hp > 0):

        print(color.RED + "\nThe Enemy Pokemon is still alive" + color.END)
        
        #Prompts User for input to choose what attack move
        move_input = input("Here are your moves:  " + color.BOLD + "Tackle " +
                           color.DARKCYAN + "  Special\n" + color.END)
        
        #After every successful attack (tackle/special) the enemy will attack
        #Also displays to the User, how much health the enemy has left and how much the user has left
        if (move_input.lower() == 'tackle'):
            
            user.tackle(enemy)
            enemy.tackle(user)
            print(color.RED + "ENEMY HP: " + str(enemy.hp) + color.END)
            print(color.BOLD + color.GREEN + "YOUR HP: " + str(user.hp) + color.END)
            
            
        elif (move_input.lower() == 'special'):
            
            user.special_attack(enemy)
            enemy.tackle(user)
            print(color.RED + "ENEMY HP: " + str(enemy.hp) + color.END)
            print(color.BOLD + color.GREEN + "YOUR HP: " + str(user.hp) + color.END)
            
        #Tests if user wants to heal(brings User's hp back to full) (Can only be used 3 Times)   
        elif (move_input.lower() == 'heal'):
            
            user.heal(counter)
            counter += 1
    
        #Tests if user wants to quit
        elif (move_input.lower() == 'end' or move_input.lower() == 'quit'):
            
            user.hp = -1
            
            
        else:
            
            print(color.RED + "Invalid Input, please try again")
            
    if (enemy.hp <= 0):         
        
        user.xp = user.xp + enemy.xp
        print(color.GREEN + "You have gained " + str(enemy.xp) + " xp" + color.END)
```

# RULES

#### You must choose between the three given starters 

#### The enemy will attack after you attack for 25 hp

#### Depending on your pokemon, you can critical hit with your special ability

#### Tackle is a 'neutral' elemant

#### You have 300 hp and you can use the heal (type 'heal') ability only (3 times) 

#### If your hp reaches (0), you lose

#### In order to win, you must kill the enemy pokemon till you reach final evolution

#### You can type 'quit' or 'end' after the selection phase, in order to quit. However, that means you lose!

```python
run()
```
